```python
def create_trade(pivot_df, perc, over=True):
    pivot_ratio_df = pivot_df.div(pivot_df.sum(axis=1), axis=0)
    mask = np.eye(pivot_ratio_df.shape[0], pivot_ratio_df.shape[1], dtype=bool)
    data = pivot_ratio_df.to_numpy()[~mask]
    threshold_percent = perc
    threshold = np.percentile(data, threshold_percent)
    if over:
        trade_overunder = pivot_ratio_df.applymap(lambda x: 1 if x>=threshold else 0)
    else:
        trade_overunder = pivot_ratio_df.applymap(lambda x:1 if x<threshold else 0)
    return trade_overunder
```

```python
trade_over80 = create_trade(pivot_df, 80, over=True)
trade_over90 = create_trade(pivot_df, 90, over=True)

trade_under20 = create_trade(pivot_df, 20, over=False)
trade_under10 = create_trade(pivot_df, 10, over=False)


trans_trade_over80 = create_trade(pivot_df.T, 80, over=True)
trans_trade_over90 = create_trade(pivot_df.T, 90, over=True)

trans_trade_under20 = create_trade(pivot_df.T, 20, over=False)
trans_trade_under10 = create_trade(pivot_df.T, 10, over=False)
```
