```python
import numpy as np
import networkx as nx

def analyze_graph(adj_matrix):
    # Create a graph from the adjacency matrix
    G = nx.Graph(adj_matrix)
    
    # Number of singletons (nodes with degree 0)
    num_singletons = sum(1 for node in G.nodes() if G.degree(node) == 0)
    
    # Number of edges
    num_edges = G.number_of_edges()
    
    # Node with the largest degree
    largest_degree_node = max(G.nodes(), key=lambda node: G.degree(node))
    largest_degree = G.degree(largest_degree_node)
    
    # Node with the least degree
    least_degree_node = min(G.nodes(), key=lambda node: G.degree(node))
    least_degree = G.degree(least_degree_node)
    
    return num_singletons, num_edges, largest_degree_node, largest_degree, least_degree_node, least_degree

def complete_analyze_graph(dataframe, print_=False):
    (ns, ne, Ldn, Ld, ldn, ld) = analyze_graph(dataframe.to_numpy())
    print(f"""number of singletons={ns}\n 
            number of edges={ne}\n
            Node w/ largest degree ={dataframe.columns[Ldn]}\n
            country name={tri2name[dataframe.columns[Ldn]]}\n
            Largest degree = {Ld}\n
            Node w/ least degree = {dataframe.columns[ldn]}\n
            country name={tri2name[dataframe.columns[ldn]]}\n
            least degree = {ld}""") if print_ else None
    return ns, ne, Ld, dataframe.columns[Ldn], tri2name[dataframe.columns[Ldn]], ld, dataframe.columns[ldn], tri2name[dataframe.columns[ldn]]
```

```python
def overleaf(networks, columns_to_show=None, networks_to_show=None):
    columns = ["n_singletons", "n_edges", "largest_degree_", "least_degree_"]
    if columns_to_show is None:
        columns_to_show = ["singletons", "edges", "degr. max", "degr. min"]
    if networks_to_show is None:
        networks_to_show = networks
    df = table.loc[networks, columns].transpose()
    n, m = df.shape
    print('& '+' & '.join(networks_to_show)) 
    print("\\clines{2-9}")
    for i in range(n):
        print(columns_to_show[i], end=' & ')
        for j in range(m):
            print(df.iloc[i, j], end='')
            print(" & ", end='') if j < m-1 else None
        print(r'\\')
        
    print()
    print(table.loc[networks, ["largest_degree_tri", "largest_degree_name", "least_degree_tri", "least_degree_name"]])
        
    
        
```
