```python
# construct the dij matrix. dij = 1 if neighbor else dmax-dij / dmax
dij = pd.DataFrame(columns=sorted(UN_countries), index=sorted(UN_countries))
dij.iloc[:, :] = float("nan")

dmax = dists.max().max()
```

```python
for tri1 in UN_countries:
    for tri2 in UN_countries:
        if df_border.loc[tri1, tri2] == True:
            dij.loc[tri1, tri2] = 1
        else:
            dij.loc[tri1, tri2] = (dmax-dists.loc[tri1, tri2]) / dmax
```
