```python
import requests
from bs4 import BeautifulSoup

# URL of the target page
url = "https://epi.yale.edu/epi-results/2022/country/nga"

# Make an HTTP GET request to fetch the page content
response = requests.get(url)
html_content = response.content

# Parse the HTML content using BeautifulSoup
soup = BeautifulSoup(html_content, 'html.parser')

# Find the table rows that match the conditions
rows = soup.find_all('tr')
start_row = None
end_row = None

for idx, row in enumerate(rows):
    if row.text.strip().startswith("EPI"):
        start_row = idx
    elif row.text.strip().startswith("GHG per capita"):
        end_row = idx
        break

# Extract data from the selected rows
if start_row is not None and end_row is not None:
    selected_rows = rows[start_row:end_row]

    # Create a dictionary to store the extracted data
    data = {}
    for row in selected_rows:
        columns = row.find_all('td')
        if len(columns) >= 2:
            key = columns[0].text.strip()
            value = columns[1].text.strip()
            data[key] = value

    # Test the validity of the extracted data
    epi_valid = data.get("EPI") == "162"
    ghg_valid = data.get("GHG per capita") == "38"

    if epi_valid and ghg_valid:
        print("Validity Test Passed:")
        print("EPI: 162")
        print("GHG per capita: 38")
    else:
        print("Validity Test Failed")

else:
    print("Table rows not found")

```

```python
from bs4 import BeautifulSoup

# Example HTML content
html_content = """
<div>
    <p>This is the first paragraph.</p>
    <p>This is the second paragraph.</p>
    <p>Search for this paragraph.</p>
    <p>Following paragraph.</p>
</div>
"""

# Create a BeautifulSoup object
soup = BeautifulSoup(html_content, 'html.parser')

# Find the first <p> tag with specific text
target_text = "Search for this paragraph."
paragraph = soup.find('p', text=target_text)

if paragraph:
    print("Found:", paragraph)
    print("Text:", paragraph.text)

    # Find the following sibling <p> tag
    following_sibling = paragraph.find_next_sibling('p')
    
    if following_sibling:
        print("Following Sibling Found:", following_sibling)
        print("Following Sibling Text:", following_sibling.text)
    else:
        print("Following Sibling not found")
else:
    print("Paragraph not found")
```
